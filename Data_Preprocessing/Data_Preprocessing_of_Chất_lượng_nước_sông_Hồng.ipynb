```python
df.drop('Unnamed: 44', axis=1, inplace=True)
df
```

```python
df.columns
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date '], format='%d/%m/%Y', errors='coerce')
```

```python
# As 'T-C' seems to represent temperature, we will try to convert it to numeric
# Any non-numeric values will be set as NaN
df['T-C'] = pd.to_numeric(df['T-C'], errors='coerce')
df
```

```python
# Impute numeric columns with median
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the aesthetics for the plots
sns.set(style="whitegrid")

# Selecting a few columns for histogram plot (example: DO, pH, Salinity)
columns_for_histogram = ['DO mgO2/l', 'pH', 'Sal ppt']

# Creating histograms for selected columns
plt.figure(figsize=(15, 5))
for i, column in enumerate(columns_for_histogram, 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Creating a box plot for one of the variables across different seasons
plt.figure(figsize=(10, 6))
sns.boxplot(x='Season', y='DO mgO2/l', data=df)
plt.title('Box Plot of Dissolved Oxygen (DO) Across Different Seasons')
plt.xlabel('Season')
plt.ylabel('Dissolved Oxygen (mgO2/l)')
plt.show()
```

```python
# prompt: box plot of ph columns according to site name, size of x axis longer

sns.boxplot(x='Site name', y='pH', data=df)
plt.title('Box Plot of pH According to Site Name')
plt.xlabel('Site Name')
plt.ylabel('pH')
plt.xticks(rotation=90)
plt.show()
```

```python
# Creating a box plot for pH across different sites
plt.figure(figsize=(15, 6))
sns.boxplot(x='Site name', y='pH', data=df)
plt.title('Box Plot of pH Across Different Sites')
plt.xlabel('Site Name')
plt.xticks(rotation=45)
plt.ylabel('pH')
plt.show()
```

```python
# Creating box plots for DO, Salinity, Nitrate, and Phosphate across different sites
plt.figure(figsize=(20, 10))

# Box plot for Dissolved Oxygen
plt.subplot(2, 2, 1)
sns.boxplot(x='Site name', y='DO mgO2/l', data=df)
plt.title('Box Plot of Dissolved Oxygen (DO) Across Different Sites')
plt.xlabel('Site Name')
plt.xticks(rotation=45)
plt.ylabel('Dissolved Oxygen (mgO2/l)')

# Box plot for Salinity
plt.subplot(2, 2, 2)
sns.boxplot(x='Site name', y='Sal ppt', data=df)
plt.title('Box Plot of Salinity Across Different Sites')
plt.xlabel('Site Name')
plt.xticks(rotation=45)
plt.ylabel('Salinity (ppt)')

# Box plot for Nitrate
plt.subplot(2, 2, 3)
sns.boxplot(x='Site name', y='NO3-N mgN/l', data=df)
plt.title('Box Plot of Nitrate (NO3-N) Across Different Sites')
plt.xlabel('Site Name')
plt.xticks(rotation=45)
plt.ylabel('Nitrate (mgN/l)')

# Box plot for Phosphate
plt.subplot(2, 2, 4)
sns.boxplot(x='Site name', y='TSP - PO4-P mgP/l', data=df)  # Assuming 'TSP - PO4-P mgP/l' as Phosphate
plt.title('Box Plot of Phosphate (PO4-P) Across Different Sites')
plt.xlabel('Site Name')
plt.xticks(rotation=45)
plt.ylabel('Phosphate (mgP/l)')

plt.tight_layout()
plt.show()
```
